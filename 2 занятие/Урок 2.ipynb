## Фильтрация столбцов в таблице
df[(Условие1)&(Условие 2)]
1. & - логическое И
1. | - логическое ИЛИ
1. ~ - логическое НЕ
1. ">" - больше 
1. "<" - меньше 
1. "==" - равно
1. df['Столбец1'].isnull() - выбрать все пустые значения в столбце 1
1. df['Столбец1'].isin(values=['Значение1','Значение2']) - значение в столбце 1 равно любому из values
1. df['Столбец1'].str.contains('Строка') - Строка является частью какого-либо значения в столбце 1
* Многие другие функции вы можете найти в документации

```python
#фильтрация 
df.loc[(df['Изделие']=='Радиоуправляемые модели машин')&(df['Неделя']=='2018-11-12')]
#можно использовать любые логические операции
```

```python
df.rename(columns={'Характеристика1': 'Бренд'}, inplace=True)
```

## Сортировка значений
df.sort_values(by=['Столбец1', 'Столбец2'], ascending=False)
* by - столбцы по которым нужно отсортировать таблицу
* ascending - сортировка по возрастанию или убыванию:
    1. False - по убыванию
    1. True - по возрастанию

```python
df.sort_values(by=['СуммаОборот'], ascending=False)
```

* sheet_name - порядковый номер листа или его название
