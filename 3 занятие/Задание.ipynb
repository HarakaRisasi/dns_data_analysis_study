# Группировка

## Задание 3
#### Сгруппируйте полученные в задании 2 данные так, чтобы по каждому Изделию был посчитан суммарный Оборот и количество уникальных Брендов

```python
df.groupby('Изделие').agg({'СуммаОборот': 'sum', 'Изделие': 'count'})
```

## * Задание 4
#### Для изделий "Смартфон" и "Телевизор" вывести Бренды, у которых суммарный оборот меньше 10 000 (по всем данным из файла)

```python
df[['Характеристика1']].loc[df['Изделие'].isin(['Смартфон', 'Телевизор'])&(df['СуммаОборот']<10000)].rename(columns={'Характеристика1': 'Бренды'})
```
