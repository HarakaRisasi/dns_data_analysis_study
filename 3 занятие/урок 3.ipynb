# Группировка

df.groupby(by=['НазваниеСтолбца1','НазваниеСтолбца2'], sort=True, as_index=True).agg({'НазваниеСтолбца3':['aggfunc1','aggfunc2'],'НазваниеСтолбца4':'aggfunc3'})

* df - исходный датафрейм
* by - список столбцов, по которым производится группировка
* sort - сортировка по столбцам группировки (True/False), по умолчанию True
* as_index - использоваться столбцы группировки как индекс (True/False), по умолчанию True
* .agg({'Столбец1':[агрегирующие функции],'Столбец2':[агрегирующие функции]}) - агрегирующие функции

**Список агрегирующих функций**:
1. 'sum' - сумма
1. 'mean' - среднее
1. 'count' - количество всех значений
1. 'unique' - уникальные значения (список)
1. 'nunique' - количество уникальных значений
1. 'max' - максимальное значение
1. 'min' - минимальное значение
1. 'std' - стандартное отклонение
1. 'median' - медиана
1. lambda x: x.value_counts().index[0] - мода ряда
1. lambda x: x.value_counts()[0] - частота моды ряда
1. любая функция, которая агрегирует данные написанная на python

### Пример 1

````python
group = df[['Операция','ДнейВРемонте','Количество']].groupby(by='Операция').mean()
group
````

### Пример 2

````python
group = df.groupby(by=['Операция','Выход']).agg({'КодТовара':['count',('Кол-во уникальных','nunique')], 'ДнейВРемонте':'mean'})
group
````

>Стоит отметить, что столбцы по которым ведется группировка - становятся индексами.

### Пример 3

````python
group = df.groupby(by=['КодТовара'], sort=False, as_index=False).agg({'ДнейВРемонте':'mean'})
group
````

> as_index=True - То колонка код товара(по ней велась группировка) становится индексом

# Сводная таблица

df.pivot_table(columns=['НазваниеСтолбца1', 'НазваниеСтолбца2'], index=['НазваниеСтолбца3', 'НазваниеСтолбца4'], values=['НазваниеСтолбца5', 'НазваниеСтолбца6'], aggfunc='sum')
* columns - столбцы сводной таблицы
* index - строки сводной таблицы
* values - значения сводной таблицы
* aggfunc - агрегирующая функция

* fill_value - заполнить NaN
* margins=True - вывести столбец и строку итогов
* margins_name - наименование строки и столбца итогов

**Список агрегирующих функций**:
1. 'sum' - сумма
1. 'mean' - среднее
1. 'count' - количество всех значений
1. 'max' - максимальное значение
1. 'min' - минимальное значение
1. 'unique' - уникальные значения (список)
1. 'nunique' - количество уникальных значений
1. 'std' - стандартное отклонение
1. 'median' - медиана
1. любая функция, которая агрегирует данные написанная на python

### Пример 1

````python
df_pivot = df.pivot_table(values='ДнейВРемонте',index='КодТовара', columns='Выход', aggfunc='mean', 
                          fill_value=0, margins=True, margins_name="Итоги")
df_pivot
````

>**Суть сводной таблицы** - это подведение итогов по определенному алгоритму (в моем случае - "mean").  
*index* - строчки в сводной таблице.  
*columns* - колонки.  
*aggfunc* - агрегирующая функция.  
*fillvalue* - заполнение NaN значений.  
*margin* and *margin name* - подсчет "итогов" по строчкам и столбцам.

> ***Чтобы ощутить разницу между сводной таблицей и операцией GroupBy***, можно представить себе сводную таблицу, как многомерный вариант агрегации посредством GroupBy.  
То есть данные разделяются, преобразуются и объединяются, но при этом разделение и объединение осуществляются не по одномерному индексу, а по двумерной сетке.

### Пример 2

````python
df_pivot = df.pivot_table(values=['ДнейВРемонте', 'КодТовара'],index=['КодФирмы'], columns=['Операция'], 
                            aggfunc={'ДнейВРемонте':'sum', 'КодТовара':'nunique'}, fill_value=0, margins=True, margins_name="Итоги")
df_pivot
````

### Пример 3

````python
df_pivot = df.pivot_table(values=['ДнейВРемонте'], index=['КодФирмы'], columns=['Выход', 'Операция'], 
                            aggfunc=[sum, 'mean', max], fill_value=0, margins=True, margins_name="Итоги")
df_pivot
````

````python
df_pivot.to_excel('pivot.xlsx')
````
