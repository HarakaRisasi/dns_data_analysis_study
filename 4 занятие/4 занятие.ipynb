## Функции. Что это? Как их использовать?

Функция в python - объект, принимающий аргументы и возвращающий значение.

Функция объявляется с помощью ключевого слово def

return - то что нужно вернуть функции

[Курс по основам Python 3](https://stepik.org/course/67/promo)

```python
def test(x):
    """
    Функция для возведения числа в квадрат
    Вход:
    x - число
    Выход:
    x ** 2
    """
    return x ** 2
```

```python
test(3)
```

```python
df['День в году'].apply(test)
```

## Получение нового столбца на основе существующей строки

* df.apply(func_name, axis=1) - применить функию ко всей таблицы

axis=1 - строка

axis=0 - столбец

```python
def con_row(x):
    """
    Удобная функция на основе которой можно маркировать соответствующие условию элементы списка.
    Данная маркировка поможет отделить "Выданый после гарантийного ремонта" товар от остальных.
    
    Вход: 
    x - операция документа
    
    Выход: 
    Маркер: "Столбец" = (1 или 0)
    """
    if x['Выход'] == 'Выдача' and x['Операция']=='Гарантийный ремонт':
        return 1
    else:
        return 0
```

```python
df['Столбец'] = df.apply(con_row, axis=1)
```
