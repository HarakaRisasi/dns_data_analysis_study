## Задание 2
#### Сгруппируйте полученные в задании 2 данные так, чтобы по каждому Году для каждого Изделия был посчитан суммарный Оборот и количество уникальных Брендов

```python
x = df.pivot_table(values=['СуммаОборот', 'Характеристика1'],
                   index=['Год', 'Изделие'], 
                   aggfunc={'СуммаОборот': 'sum', 'Характеристика1': 'nunique'},
                   margins=True, 
                   margins_name="Итоги:").rename(columns={'Характеристика1' :'Кол-во уникальных брендов'})
                   
x
```

> Экспоненциальный формат отображает число в экспоненциальном представлении, заменяя часть числа на e +n, в которой e (экспонента) умножает предшествующее число на $10$ в $n-ом$. Например, в экспоненциальном формате с двумя десятичными форматами $12345678901$ в качестве $1.23 E + 10$, что $1,23$ подает $10$ в десятый источник.

```python
# В данном задании неполучается использовать сводную таблицу, так как "aggfunc ='sum' or ['sum', unique]", будет применяться ко всем колонкам, а это не правильно.
# Поэтому более разумно здесь использовать функцию "groupby".
group = df.groupby(['Год', 'Изделие']).agg({'СуммаОборот': 'sum', 'Характеристика1': 'nunique'})
group.rename(columns={'Характеристика1': 'Кол-во уникальных'})
```

## Задание 5
#### Создайте новый столбец исходя из следующих условий: если выход выдачи или закрытие заказа, то ОСНТ, иначе АД. Посчитайте количество различных документов по каждому полученном условию

```python
val_list = df['Выход'].unique().tolist()[0::2]
val_list
```

```python
def condition(x):
    if x in  val_list:
        return 'ОсНТ'
    else:
        return 'АД'
```

```python
df['Решение'] = df['Выход'].apply(condition) # "axis" не нужен так как применяется к Series.
df.head()
```
