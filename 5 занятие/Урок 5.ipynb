## Соединения таблиц<a class="anchor" id="bullet-15"></a>
df.merge(right=product, left_on=['Столбец1ЛевойТаблицы','Столбец2ЛевойТаблицы'], right_on=['Столбец1ПравойТаблицы','Столбец2ПравойТаблицы'], how='inner')
* df - таблица слева
* right - таблица справа
* left_on - ключ для соединения из левой таблицы
* right_on - ключ для соединения из правой таблицы
* how - как будет произведенно соединение:
    1. inner - использовать только пересечение ключей из левой и правой таблицы, аналогично внутреннему соединению в SQL
    1. left - использовать ключи из левой таблицы, аналогично левому соединению в SQL
    1. right - использовать ключи из правой таблицы, аналогично правому соединению в SQL
    1. outer - использовать объединение ключей из обоих таблиц, аналогично полному или внешнему соединению из SQL
