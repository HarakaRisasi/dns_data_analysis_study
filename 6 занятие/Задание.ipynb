```python
import pandas as pd
import plotly.express as px
```

# Загрузите табличку "Сроки ремонтов" из 3 занятия

```python
df_repair_terms = pd.read_csv('../3 занятие/Сроки ремонта.csv')
df_info_goods = pd.read_csv('Инф_товар.csv')
```

# Присоединить к таблице срок ремонтов информацию о товарах

```python
df_repair_terms.head(2)
```

```python
df_info_goods.head(2)
```

```python
df_ = df_repair_terms.merge(right=df_info_goods, left_on='КодТовара', right_on='Код', how='right')
df_ = df_.drop(columns='Код')
```

```python
df_.isnull().sum()
```

```python
df_
```

# Посчитать средний срок ремонтов по каждому изделию

```python
df_mean = df_.pivot_table(values='ДнейВРемонте',index='Изделие', aggfunc='mean')
df_mean
```

# Вывести первые 20 по убыванию срока ремонтов изделие

```python
df_mean = df_mean[:20].sort_values(by='ДнейВРемонте', ascending=False).reset_index()
df_mean
```

# Построить столбчатую диаграмму

```python
fig = px.bar(
    df_mean,
    x='Изделие',
    y='ДнейВРемонте',
    color='Изделие',
    title='Исделия с наибольшим сроком ремонта'
)
fig.show()
```
